```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
pd.options.display.max_rows = 1000
```

```python
df = pd.read_excel('train_data_2023.xlsx')
```

```python
# 이상치 제거 및 다회 구매 고객 Filtering
pivot = df.pivot_table(index = 'Customer_ID', values = 'Order_Date', aggfunc = 'count' )
df0 = df[df['Customer_ID'].isin(pivot[(pivot['Order_Date'] < 10) & (pivot['Order_Date'] > 3)].index)].tail(30000)
```

```python
# 데이터 전처리
df2 = df0.groupby(['Order_Date', 'Customer_ID'])['Product_No'].apply(list).reset_index()
te = TransactionEncoder()
te_ary = te.fit(df2['Product_No']).transform(df2['Product_No'])
df2 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Apriori 분석
frequent_itemsets = apriori(df2, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```
